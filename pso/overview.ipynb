```python
import numpy as np
```

```python
import pandas as pd
import pickle
```

```python
data51 = pickle.load(open('pso50100','rb'))
data55 = pickle.load(open('pso50500','rb'))
data21 = pickle.load(open('pso200100','rb'))
data25 = pickle.load(open('pso200500','rb'))
```

```python
data = {
    'N=050 E=100':data51,
    'N=050 E=500':data55,
    'N=200 E=100':data21,
    'N=200 E=500':data25,
}
```

```python
dados = []
for key,value in data.items():
    for alg,alg_data in value.items():
        if alg == 'pso_fr' or alg == 'pso_': continue
        algoritmo = [key, alg.upper() if alg != 'iwpso' else 'PSO']
        for dim, dim_data in alg_data.items():
            algoritmo.append(dim_data['mean'])
        dados.append(algoritmo)
```
